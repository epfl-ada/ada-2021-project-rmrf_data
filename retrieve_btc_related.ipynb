```python
import pandas as pd
import os

from pathlib import Path
```

```python
# Read data from zip file, set chunksize to fit data into RAM memory 
FILEPATH = '/Users/xuyixuan/Downloads/quotes-2017.json.bz2'

# Initialise a list of keywords to filter the data
KEYWORDS = ["bitcoin", "BTC", "crypto asset", "crypto currenc", "cryptos", "crypocurrenc"]
```

```python
def preprocess_data(data_chunk):

    # Filter all data chunks with selected keywords
    data_chunk = data_chunk[data_chunk["quotation"].str.contains('|'.join(KEYWORDS), case=False)]

    return data_chunk

def export_df():
    data = pd.read_json(FILEPATH, lines=True, compression='bz2', chunksize=1000000)
    chunk_list = []
    
    for chunk in data:
        chunk_list.append(preprocess_data(chunk))
    
    # Concat all data chunks to a single dataframe
    df_concat = pd.concat(chunk_list)
    
    # Generate corresponding csv file name
    csv_name = "{}-BTC.csv".format(os.path.basename(FILEPATH).split('.')[0])
    
    # Output csv file
    df_concat.to_csv("./data/" + csv_name, index=False)
```

```python
# Export one specific dataframe to csv file
# Team members run this code in different zipped files to save time
export_df()
```

```python
# Retrieve a list of generated csv files in the data folder
PATH = "./data/"
files = Path(PATH).rglob("*-BTC.csv")

# Read the csv files into a list of dataframes
all_csvs = [pd.read_csv(file) for file in files]

# Concat pandas dataframes into one
all_csvs = pd.concat(all_csvs)

# Export the caoncat dataframe to csv to facilitate later use
all_csvs.to_csv("./data/quotes-BTC.csv", index=False)
```
